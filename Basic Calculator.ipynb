```python
# Basic Calculator using Function
def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def multi(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
def rem(a,b):
    return(a%b)
def flo(a,b):
    return(a//b)
print(" Select the apporiate operations")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Divison")
print("5.Reminder")
print("6. Floor division")
while True:
    a= float(input(" Enter the frist number: "))
    b= float(input(" Enter the second number: "))
    c= int(input("Enter the apporicate operations: "))
    if c in (1,2,3,4,5,6,):
        if c==1:
            print("Addition :   ",add(a,b))
        elif c==2:
            print("Subtraction: ", sub (a,b))
        elif c==3:
            print(" Multiplication:   ",multi(a,b ))
        elif c==4:
            print(" Division :   ", div (a,b))
        elif c==5:
            print("Reminder :  " , rem (a,b))
        elif c==6:
            print(" Floor division : ", flo (a,b))
        else:
            print("Invilid Number")
        di= input("\n Enter operation can be done ? Y/N")
        if di =="N":
            print("OFF")
            break
    else:
        print("Invilid input")
    
        
        
    
    
```
